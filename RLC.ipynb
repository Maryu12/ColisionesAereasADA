```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
```

```python
# parametros del circuito
Vs = 10
R = 10
L = 0.01
C = 100e-6
```

```python
# funcion de transferencia I(s)/Vs
num = [1]
den = [L, R, 1/C]
```

```python
system = signal.TransferFunction(num, den)
```

```python
# respuesta al escalon
t, y = signal.step(system)
```

```python
# multiplicar por Vs
i = Vs*y
```

```python

plt.figure()
plt.plot(t, i)
plt.xlabel("Tiempo (s)")
plt.ylabel("Corriente i(t) (A)")
plt.title("Respuesta de corriente en circuito RLC")
plt.grid()
plt.show()
```
